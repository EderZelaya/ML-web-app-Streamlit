```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# RUTA ABSOLUTA donde ya sabemos que está el dataset_split
BASE = "/workspaces/Eder_Zelaya-Machine-Learnig/data/raw/dogs-vs-cats/dataset_split"
TRAIN_DIR = os.path.join(BASE, "train")
VALID_DIR = os.path.join(BASE, "validation")

print("BASE:", BASE)
print("TRAIN_DIR existe?", os.path.isdir(TRAIN_DIR))
print("VALID_DIR existe?", os.path.isdir(VALID_DIR))

IMG_SIZE = (224, 224)
BATCH_SIZE = 32

if not os.path.isdir(TRAIN_DIR):
    raise FileNotFoundError(f"No existe TRAIN_DIR: {TRAIN_DIR}")
if not os.path.isdir(VALID_DIR):
    raise FileNotFoundError(f"No existe VALID_DIR: {VALID_DIR}")

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

valid_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True,
    seed=42
)

valid_generator = valid_datagen.flow_from_directory(
    VALID_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    seed=42
)

print("Train samples:", train_generator.samples)
print("Valid samples:", valid_generator.samples)
print("Class indices:", train_generator.class_indices)
```

```python
import os
import pandas as pd

BASE = "/workspaces/Eder_Zelaya-Machine-Learnig/data/raw/dogs-vs-cats/dataset_split"

paths = []
labels = []

# Recorremos las dos carpetas: train y validation
for subset in ["train", "validation"]:
    for label in ["cats", "dogs"]:
        folder = os.path.join(BASE, subset, label)
        for filename in os.listdir(folder):
            if filename.lower().endswith((".jpg", ".jpeg", ".png")):
                paths.append(os.path.join(folder, filename))
                labels.append(label)

df = pd.DataFrame({"filepath": paths, "label": labels})

df.head()
```
